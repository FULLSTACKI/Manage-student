```python
import pandas as pd
from sqlite3 import *
from sqlalchemy import *
```

```python
engine = create_engine("sqlite:///../student_score.db")
query = """
        SELECT 
                    st.student_id,
                    st.student_name,
                    st.birthday,
                    sc.gpa,
                    d.department_name
                FROM students as st 
                JOIN scores as sc ON sc.student_id = st.student_id 
                JOIN departments as d ON d.department_id = st.department_id
                ORDER BY sc.gpa DESC 
                LIMIT 3 
"""
result_df = pd.read_sql(query, engine)
result_df
```
